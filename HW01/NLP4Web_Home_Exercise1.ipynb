Tokenization, also known as word segmentation, is the task of splitting a stream of characters into a ordered list of semantically useful groups of characters such as words and punctuation (e.g. the sentence "The cat sat on the mat." consists of tokens "The", "cat", "sat", "on", "the", "mat", "."). For the purpose of this task, you can assume that a token should be separated whenever a space (" ") or punctuation (".", ",", ";", "!") is encountered. Tokenization is a vital part of the NLP pipeline as it enables downstream linguistic analysis of text (e.g. how many nouns are in the text ?).

a) Implement the function `word_tokenize` in the provider Jupyter notebook and tokenize the above tweet. <s>Use pure Python for this exercise (e.g. no internal or external libraries such as `re`)</s>. **EDIT:** In today's Practice class I explicitly allowed the use of internal python libraries.
```python
tweet = "    While we're waiting for @LeagueOfLegends's 10th anniversary. Let's have some fun! Tweet me your favorite skin ideas and I'll send 1k RP to a couple of my favorites. My friend keeps telling me he'd like to see ShamWow! Vlad. #League10    "
t ="The cat sat on the mat."

def word_tokenize(text):
    """
    Segments a stream of characters into tokens
    
    Example: "The cat sat on the mat." -> ["The", "cat", "sat", "on", "the", "mat", "."]
    :param text: String that should be tokenized
    :return: List of token strings
    """
    token_list = []
    delimiters = ['#','.',',',';','!',' ','@',"'"]
    indexes= []
    
    for index, char in enumerate(text):
        if char in delimiters: 
            indexes.append(index)
    
    pointer = 0 
    for i in indexes: 
        if text[pointer:i]!= '' and text[pointer:i]!= ' ':
            token_list.append(text[pointer:i])
        if text[i]!= ' ' and text[i] != '':
            token_list.append(text[i])
        pointer = i + 1 
    
    return token_list

tokenized_tweet = word_tokenize(tweet)
print(tokenized_tweet)
```

b) Print a list of all tokens and their length. Each token must only be printed once.

```python
for token in tokenized_tweet: 
    print("Token: {} .Length {}".format(token, len(token)))
```

c) Print a list of all tokens and their frequency. Each token must only be printed once.

```python
freq_dict = {}

for token in tokenized_tweet:
    if token not in freq_dict: 
        freq_dict[token] = 1
    else:
        freq_dict[token] += 1

print(freq_dict)
```
